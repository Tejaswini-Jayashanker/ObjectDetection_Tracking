**HAAR CASCADE Classifier - Human Face, Human eyes detection**

```python
#Library imports
import cv2
from google.colab.patches import cv2_imshow
import numpy
```

```python
#Load my selfie to perform the Face and Eye detection and draw a Rectangle/Circle:-
selfie = cv2.imread('/content/me.jpg')
cv2_imshow(selfie)
```

```python
#Everything is in CV2 now, no need to download the XML files,
#just give the path as mentioned in OPENCV Github, "OpenCV/data/Cascades"
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades  + 'haarcascade_eye.xml')
```

```python
#Convert to GrayScale as the algorithm faster:
gray_image = cv2.cvtColor( selfie, cv2.COLOR_BGR2GRAY )

#Store the Faces identified
faces = face_cascade.detectMultiScale(gray_image, 1.3, 5)
if faces is ():
    print("No faces found")
else:
    for (x,y,w,h) in faces:
        cv2.rectangle(selfie ,(x,y),(x+w,y+h),(127,0,255),2)
        roi_gray = gray_image[y:y+h, x:x+w]
        roi_color = selfie[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.2, 5)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,255,0),2)
cv2_imshow(selfie)
```

Okay, I will try smile detection now:

```python
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')
selfie = cv2.imread('/content/me.jpg')
gray_image = cv2.cvtColor( selfie, cv2.COLOR_BGR2GRAY )
face = face_cascade.detectMultiScale( gray_image, 1.3, 5)
if face is ( ):
    print("No face Found!")
else:
    for (x,y,w,h) in face:
        cv2.rectangle( selfie, (x,y), (x+w, y +h), (127,0,255), 3 )
        roi_gray = gray_image[y:y+h, x:x+w]
        roi_color = selfie[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 22)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,255,0), 2)
        smiles = smile_cascade.detectMultiScale( roi_gray, 1.7, 22 )
        for (sx, sy, sw, sh ) in smiles:
            cv2.rectangle(selfie, (sx, sy), (sx+sw, sy+sh),  (0, 0, 255), 2 )
cv2_imshow(selfie)
```
